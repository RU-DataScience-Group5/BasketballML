```python
import pandas as pd

# set up files as an array of filename and season
Files = [["Data/AdvData/adv05_06.csv","2005-06"],
         ["Data/AdvData/adv06_07.csv","2006-07"],
         ["Data/AdvData/adv07_08.csv","2007-08"],
         ["Data/AdvData/adv08_09.csv","2008-09"],
         ["Data/AdvData/adv09_10.csv","2009-10"],
         ["Data/AdvData/adv10_11.csv","2010-11"],
         ["Data/AdvData/adv11_12.csv","2011-12"],
         ["Data/AdvData/adv12_13.csv","2012-13"],
         ["Data/AdvData/adv13_14.csv","2013-14"],
         ["Data/AdvData/adv14_15.csv","2014-15"],
         ["Data/AdvData/adv15_16.csv","2015-16"],
         ["Data/AdvData/adv16_17.csv","2016-17"],
         ["Data/AdvData/adv17_18.csv","2017-18"],
         ["Data/AdvData/adv18_19.csv","2018-19"]
        ]

# Create empty dataframe
df_AdvancedStats = pd.DataFrame()
```

```python
# Loop through each file and append to df_AdvancedStats dataframe
for file in Files:
    df = pd.read_csv(file[0])
    df["season"] = file[1]
    df["PlayerName"] = df["Player"].str.split("\\", n=1, expand=True)[0]
    df["PlayerID"] = df["Player"].str.split("\\", n=1, expand=True)[1]
    df_AdvancedStats = df_AdvancedStats.append(df)
    
df_AdvancedStats.reset_index(drop=True,inplace=True)

# Drop unnecessary columns
df_AdvancedStats.drop(columns=['Unnamed: 19','Unnamed: 24'],inplace=True)
df_AdvancedStats
```
