```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
mvp = pd.read_csv("Data/MVP.csv")
# mvp = mvp.drop("Unnamed: 0", axis=1)
mvp
```

```python
mvp.dtypes
```

```python
mvp.astype({"Age": "float64"}).dtypes
```

```python
mean_mvp = mvp["WS"].mean()
print(f"Average Win Share by MVP candidate: {mean_mvp}")
```

```python
X = mvp.drop(["Rank", "Player", "Tm", "First", "Pts Won", "Pts Max", "Share", "WS", "season"], axis=1)
y = mvp["WS"]
print("Shape: ", X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

predicted = model.predict(X_test)

# mean squared error and r-squared score
mse = mean_squared_error(y_test, predicted)
r2 = r2_score(y_test, predicted)

print(f"MSE: {mse}")
print(f"R2: {r2}")
```
